# Launch the Syft Enclave Runner

Runs the enclave as its own process — `python -m syft_enclaves`, configured by
`SYFT_ENCLAVE_*` environment variables (see `syft_enclaves.settings.EnclaveSettings`).

This notebook is **dedicated to the enclave**: the run cell below runs the runner
in the foreground — it blocks and streams the runner's logs inline. That is
intended. To stop it, **interrupt the kernel** (Kernel → Interrupt); the runner
catches the signal and shuts down gracefully.

Run this notebook first, then run `enclave_eval_e2e.ipynb` from a separate kernel.

## Configuration

```python
import os
import sys
from pathlib import Path

# Load .env from this folder (same pattern as test_email_e2e.ipynb).
ENV_PATH = Path(".env")
assert ENV_PATH.exists(), (
    f".env not found at {ENV_PATH.resolve()} — copy .env.example to .env and fill it in"
)
for line in ENV_PATH.read_text().splitlines():
    line = line.strip()
    if line and not line.startswith("#"):
        key, _, value = line.partition("=")
        os.environ.setdefault(key.strip(), value.strip())

ENCLAVE_EMAIL = os.environ["SYFT_ENCLAVE_EMAIL"]
ENCLAVE_TOKEN = Path(os.environ["SYFT_ENCLAVE_TOKEN_PATH"])

print(f"Enclave email : {ENCLAVE_EMAIL}")
print(f"Token path    : {ENCLAVE_TOKEN}")
print(f"Token exists  : {ENCLAVE_TOKEN.exists()}")
print(f"Poll interval : {os.environ.get('SYFT_ENCLAVE_POLL_INTERVAL', '10')}s")
print(f"Require TEE   : {os.environ.get('SYFT_ENCLAVE_REQUIRE_TEE', 'false')}")
```

## Run the enclave runner

The cell below runs `python -m syft_enclaves` in the **foreground** — it blocks
and streams the runner's logs here. Leave it running.

**To stop:** interrupt the kernel (Kernel → Interrupt). The runner handles the
signal and shuts down cleanly.

```python
# This blocks — the runner is an infinite poll loop. Interrupt the kernel to stop.
assert ENCLAVE_TOKEN.exists(), (
    f"Enclave token missing at {ENCLAVE_TOKEN} — create it with the cell above first"
)
print(f"Starting `python -m syft_enclaves` for {ENCLAVE_EMAIL}")
print("(this cell blocks; interrupt the kernel to stop)")
print()

!{sys.executable} -m syft_enclaves
```
