Delete each peer's SyftBox before launching the enclave. Run the cells in order — each pops a Google login (pick the matching account), deletes that peer's SyftBox, then logs out so the next cell can sign in as a different account.

```python
import os, shutil, subprocess
os.environ["PRE_SYNC"] = "false"

from syft_enclaves import login_do, login_ds

ENCLAVE_EMAIL = "test.enclave@gmail.com"
MODEL_OWNER_EMAIL = "test.model.owner@gmail.com"
BENCHMARK_OWNER_EMAIL = "test.benchmark.owner@gmail.com"
RESEARCHER_EMAIL = "test.researcher@gmail.com"

def logout():
    subprocess.run(["gcloud", "auth", "revoke", "--all"], capture_output=True)
    shutil.rmtree("/root/.config/gcloud", ignore_errors=True)
    shutil.rmtree(os.path.expanduser("~/.config/gcloud"), ignore_errors=True)
    print("logged out")
```

```python
# Model owner — sign in as MODEL_OWNER_EMAIL
model_owner = login_do(email=MODEL_OWNER_EMAIL, sync=False, load_peers=False)
model_owner._manager.delete_syftbox()
logout()
```

```python
# Benchmark owner — sign in as BENCHMARK_OWNER_EMAIL
benchmark_owner = login_do(email=BENCHMARK_OWNER_EMAIL, sync=False, load_peers=False)
benchmark_owner._manager.delete_syftbox()
logout()
```

```python
# Researcher — sign in as RESEARCHER_EMAIL
researcher = login_ds(email=RESEARCHER_EMAIL, sync=False, load_peers=False)
researcher._manager.delete_syftbox()
logout()
```
